```python
image = Image.open("suo.jpg")
width, height = (300, 200)
new_size = (width, height)
resized_image = image.resize(new_size)
resized_image.save("suo_resized.jpg")
resized_image
```

```python
from PIL import Image
image = Image.open("suo_resized.jpg")
resized_image.save("suo_resized.jpg")
crop_area = (10, 10, 10, 10) # įrašykite norimus parametrus, pvz.: (100, 100, 300, 200)
cropped_image = image.crop(crop_area)
cropped_image.save("suo_resized.jpg")
```
